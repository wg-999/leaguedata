```python
def get_matches(region,puuid,q_type,count,api_key):
    api_url = f"https://{region}.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?type={q_type}&start=0&count={count}&api_key={api_key}"

    resp = requests.get(api_url)

    return resp.json()


```

```python
def get_all_match_data_normalized(match_ids, player_puuid, api_key, region) :
    """
    grabbing the data from each of our matches(last 50) to make a dataframe for further analysis. For now, just grabbing literally everything. 
    I will clean it later or trim the function to grab relevant stuff.
    """
    
    base_url = f"https://{region}.api.riotgames.com/lol/match/v5/matches/"
    headers = {"X-Riot-Token": api_key}
    
    all_player_data = []
    
    for i, match_id in enumerate(match_ids):
        try:
            print(f"Fetching match {i+1}/{len(match_ids)}: {match_id}")
            
            
            response = requests.get(f"{base_url}{match_id}", headers=headers)
            
            if response.status_code == 200:
                match_json = response.json()
                
                #getting the player we want 
                metadata = match_json['metadata']
                info = match_json['info']
                
                try:
                    participant_index = metadata['participants'].index(player_puuid)
                    player_data = info['participants'][participant_index]
                    
                    player_data['match_id'] = metadata['matchId']
                    player_data['game_duration'] = info['gameDuration']
                    player_data['game_mode'] = info['gameMode']
                    player_data['game_type'] = info['gameType']
                    player_data['queue_id'] = info['queueId']
                    player_data['game_version'] = info['gameVersion']
                    player_data['game_creation'] = info['gameCreation']
                    player_data['game_end'] = info.get('gameEndTimestamp')
                    
                    all_player_data.append(player_data)
                    
                except ValueError:
                    print(f"Player not found in match {match_id}")
                   #i had to add sleeping to this bc i can only do so many requests on trial api key.  
            elif response.status_code == 429:
                print("Rate limited. Waiting 2 seconds...")
                time.sleep(2)
                continue
            else:
                print(f"Error fetching {match_id}: {response.status_code}")
                
        except Exception as e:
            print(f"Exception for match {match_id}: {e}")
            
        
        time.sleep(0.1)
    
    # normalize the json data to a df 
    if all_player_data:
        df = pd.json_normalize(all_player_data)
        print(f"\nDataFrame created with {len(df)} matches and {len(df.columns)} columns!")
        print(f"Shape: {df.shape}")
        return df
    else:
        print("No match data collected")
        return pd.DataFrame()
```
